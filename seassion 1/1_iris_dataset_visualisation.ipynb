La méthode describe() renvoie les statistiques descriptives suivantes: 

- Comptage (Count) : Le nombre d'éléments non manquants dans la colonne.<br>
- Moyenne (Mean) : La moyenne des valeurs dans la colonne.<br>
- Écart-type (Standard Deviation - std) : L'écart-type des valeurs dans la colonne, mesurant la dispersion des données par rapport à la moyenne.<br>
- Minimum : La valeur minimale dans la colonne.<br>
- Quartiles (25%, 50%, 75%) : Les valeurs qui divisent les données en quatre parties égales. Le quartile à 50% est également appelé la médiane.<br>
- Maximum : La valeur maximale dans la colonne.<br>

N.B : L'écart-type (std) est une mesure importante de la dispersion des données. Plus l'écart-type est élevé, plus les valeurs sont dispersées par rapport à la moyenne, ce qui peut indiquer une plus grande variabilité dans les données.

```python
plt.plot(dataset["sepal_length"], 'o')
plt.xlabel('Le numéro de la plante ', fontsize=12, color='blue')
plt.ylabel('La longeur de la sépale ', fontsize=12, color='red')
plt.show()
```

```python
plt.plot(dataset["sepal_width"], 'o')
plt.xlabel('Le numéro de la plante ', fontsize=12, color='blue')
plt.ylabel('La largeur de la sépale ', fontsize=12, color='red')
plt.show()
```

```python
plt.plot(dataset["sepal_length"], dataset["sepal_width"], 'o')
plt.xlabel('La largeur de la sépale ', fontsize=12, color='blue')
plt.ylabel('La longueur de la sépale ', fontsize=12, color='red')
plt.show()
```
